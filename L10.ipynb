```
#mision_1
zones = (gpd.read_file('/zonas.shp',crs={'init':'epsg:4326'}))
display(zones.head(5))
zones.plot()
```

```
#misión_2
parades = (gpd.read_file('/paradas.shp',crs={'init':'epsg:4326'}))
display(parades.head(5))
parades.plot()
```

```

grafico = zones.plot(color='white', edgecolor='black')

parades.plot(ax=grafico, marker='o', color='red', markersize=5);
```

```
#mision_3
graf = gpd.sjoin(parades, zones)
grafico = zones.plot(color='white', edgecolor='black')
graf.plot(ax=grafico, marker='o', color='red', markersize=5);
```
